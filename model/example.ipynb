```python
#gene100
#CNN-Cox
#types
conv1_size=(1, 10)
conv2_size=(10, 1)
input_shape = (10, 10, 1)
le, wi = 10,10
seed=42
cancer_types= ['BLCA','HNSC','KIRC','LGG','LUAD','LUSC','SKCM']
Cindex=[]
for cancer in range(len(cancer_types)):
    #path
    #current_path='.../CNN-Cox/data/100/'
    current_path='../data/100'
    cancer_name = cancer_types[cancer]
    print(cancer_name)
    output_path=current_path+'/' + cancer_name + "100.csv"
    
    #save_path=".../CNN-Cox/save model/100/CNN-Cox/"
    save_path="../save model/100/CNN-Cox/"
    if cancer_name=="BLCA":
        conv1=128
        conv2=32
        dense=64
    if cancer_name=='HNSC':
        conv1=8
        conv2=32
        dense=16
    if cancer_name=='KIRC':
        conv1=16
        conv2=8
        dense=16
    if cancer_name=='LGG':
        conv1=128
        conv2=8
        dense=32
    if cancer_name=='LUAD':
        conv1=16
        conv2=32
        dense=32
    if cancer_name=='LUSC':
        conv1=128
        conv2=64
        dense=16
    if cancer_name=='SKCM':
        conv1=32
        conv2=32
        dense=16 
    score_tst_list = traincnncoxmodel(output_path,cancer_name,conv1,conv1_size,conv2,conv2_size,dense,input_shape,save_path,le, wi,seed) 
    Cindex.append(score_tst_list)
numbers=[422,543,596,514,551,539,454]
avgci = avgcindex(Cindex,cancer_types,numbers)
print("Micro-average C-index:",avgci)   
```
